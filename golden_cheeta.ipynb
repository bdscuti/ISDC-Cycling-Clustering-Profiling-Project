Calculates the Number of Days Between Oldest and Most Recent Workout for all athletes from locally stored athletes and their activities.

```python
import os
import json
from datetime import datetime

data_dir = "./Golden_Cheeta/data"

def get_ride_dates_from_json(json_file_path):
    try:
        with open(json_file_path, "r") as f:
            data = json.load(f)
    except json.JSONDecodeError as e:
        print(f"Error decoding JSON in file {json_file_path}: {e}")
        return None, None

    rides = data.get("RIDES", [])
    if not rides:
        return None, None

    first_date_str = rides[0]["date"]
    last_date_str = rides[-1]["date"]
    dt_format = "%Y/%m/%d %H:%M:%S %Z"
    try:
        first_date = datetime.strptime(first_date_str, dt_format)
        last_date = datetime.strptime(last_date_str, dt_format)
    except ValueError as e:
        print(f"Error parsing dates in file {json_file_path}: {e}")
        return None, None

    return first_date, last_date

results = {}
for rider_id in os.listdir(data_dir):
    rider_folder = os.path.join(data_dir, rider_id)
    if os.path.isdir(rider_folder):
        num_activities = 0
        overall_earliest = None
        overall_latest = None

        for filename in os.listdir(rider_folder):
            if filename.endswith(".json"):
                num_activities += 1
                file_path = os.path.join(rider_folder, filename)
                earliest, latest = get_ride_dates_from_json(file_path)
                if earliest and latest:
                    if overall_earliest is None or earliest < overall_earliest:
                        overall_earliest = earliest
                    if overall_latest is None or latest > overall_latest:
                        overall_latest = latest

        if overall_earliest and overall_latest:
            day_diff = (overall_latest - overall_earliest).days
            ratio = num_activities / day_diff if day_diff != 0 else None
        else:
            day_diff = 0
            ratio = None

        results[rider_id] = {
            "days_difference": day_diff,
        }


# Stored in days_difference
days_difference = pd.DataFrame(results).T["days_difference"]
```

```python
import os
import pandas as pd

# Path to the base directory
base_dir = "Golden_Cheeta/data"
# List to store person IDs with 250 < file count < 1000
valid_ids = []
# Dictionary to store {person_id: number_of_activities}
activity_counts = {}
# Loop through each subfolder in the base directory
for person_id in os.listdir(base_dir):
    person_path = os.path.join(base_dir, person_id)
    # Check if it's a directory
    if os.path.isdir(person_path):
        # Count the number of files (excluding subdirectories)
        file_count = sum(
            1 for item in os.listdir(person_path)
            if os.path.isfile(os.path.join(person_path, item))
        )
        # Append the ID and update dictionary if file count is within range
        if lower_bound < file_count < upper_bound:
            valid_ids.append(person_id)
            activity_counts[person_id] = file_count
# Convert to DataFrame correctly
activity_counts = pd.DataFrame(list(activity_counts.items()), columns=["Person_ID", "Activity_Count"])
activity_counts.set_index("Person_ID", inplace=True)
# Output the result
print("Person IDs with 250 < files < 3000:", valid_ids)
print(activity_counts)
```

Code for Aggregating all activities per athlete. Compute as new columns the mean, std, min, and max. Final Dataframe gets stored as rider_summary_2.csv

```python
import pandas as pd 
import numpy as np
from opendata import OpenData
from collections import defaultdict
from concurrent.futures import ThreadPoolExecutor, as_completed

def safe_float(val):
    if isinstance(val, list):
        val = val[0]
    try:
        return float(val)
    except Exception:
        return np.nan

# Clean allowed athlete IDs (remove any curly braces)
allowed_athlete_ids = {athlete_id.strip("{}") for athlete_id in activity_between_250_and_5000_athletes}

number_of_athletes_processed = 0

def process_athlete(athlete):
    global number_of_athletes_processed
    try:
        # Get the athlete ID from metadata and clean it
        raw_id = athlete.metadata.get('ATHLETE', {}).get('id', 'Unknown')
        athlete_id = raw_id.strip("{}")
        print(f"Processing athlete: {athlete_id}")
        number_of_athletes_processed += 1
        if number_of_athletes_processed % 100 == 0:
            print(f"Processed {number_of_athletes_processed} athletes so far in processing...")
            
        athlete_meta = athlete.metadata.get('ATHLETE', {})
        gender = athlete_meta.get('gender')
        yob = int(athlete_meta.get('yob', 0))

        metric_accumulator = defaultdict(list)
        activity_count = 0

        activities = list(athlete.activities())
        for activity in activities:
            metadata = activity.metadata
            if not metadata:
                continue
            metrics = metadata.get('METRICS', {})
            if not metrics:
                continue

            activity_count += 1
            for key, value in metrics.items():
                num = safe_float(value)
                if not np.isnan(num):
                    metric_accumulator[key].append(num)

        if activity_count == 0:
            return None

        athlete_summary = {
            'athlete_id': athlete_id,
            'gender': gender,
            'yob': yob,
            'activity_count': activity_count
        }

        for metric, values in metric_accumulator.items():
            values_np = np.array(values, dtype=np.float32)
            athlete_summary[f'{metric}_mean'] = np.nanmean(values_np)
            athlete_summary[f'{metric}_std'] = np.nanstd(values_np)
            athlete_summary[f'{metric}_min'] = np.nanmin(values_np)
            athlete_summary[f'{metric}_max'] = np.nanmax(values_np)

        return athlete_summary

    except Exception as e:
        print(f"Error processing athlete: {e}")
        return None

def process_in_batches(athletes, batch_size=100, max_workers=8):
    all_summaries = []
    total = len(athletes)
    for i in range(0, total, batch_size):
        print(f"Processing batch {i // batch_size + 1} of {((total - 1) // batch_size) + 1}")
        batch = athletes[i:i+batch_size]
        with ThreadPoolExecutor(max_workers=max_workers) as executor:
            futures = [executor.submit(process_athlete, a) for a in batch]
            for future in as_completed(futures):
                result = future.result()
                if result:
                    all_summaries.append(result)
    return all_summaries

# Main
od = OpenData()

# Fetch only the athletes in the allowed list using cleaned IDs.
athletes = []
for athlete_id in allowed_athlete_ids:
    try:
        athlete = od.get_local_athlete(athlete_id)
        athletes.append(athlete)
        print(f"Fetched athlete: {athlete_id}")
    except Exception as e:
        print(f"Failed to fetch athlete {athlete_id}: {e}")

print(f"Total athletes fetched: {len(athletes)}")

summaries = process_in_batches(athletes, batch_size=100, max_workers=16)

df = pd.DataFrame(summaries)
df.to_csv('rider_summary_2.csv', index=False)
print("Processing complete. Summary saved to rider_summary_2.csv")
```

```python
import numpy as np

# Your list of column names
critical_power_mean_list = [
    '1s_critical_power_mean',
    '5s_critical_power_mean',
    '10s_critical_power_mean',
    '20s_critical_power_mean',
    '30s_critical_power_mean',
    '1m_critical_power_mean',
    '2m_critical_power_mean',
    '3m_critical_power_mean',
    '5m_critical_power_mean',
    '20m_critical_power_mean',
    '60m_critical_power_mean',
    '90m_critical_power_mean'
]

# Corresponding durations in seconds
durations = np.array([
    1,
    5,
    10,
    20,
    30,
    60,        # 1m
    2 * 60,    # 2m
    3 * 60,    # 3m
    5 * 60,    # 5m
    20 * 60,   # 20m
    60 * 60,   # 60m
    90 * 60    # 90m
])

# Create slope feature names between adjacent points
slope_feature_names = []
for i in range(len(critical_power_mean_list) - 1):
    start = critical_power_mean_list[i].replace('_mean', '')
    end = critical_power_mean_list[i+1].replace('_mean', '')
    slope_feature_names.append(f"slope_{start}_to_{end}")

# Function to compute local slopes
def compute_local_slopes(row):
    y = row[critical_power_mean_list].values.astype(float)
    slopes = []
    for i in range(len(durations) - 1):
        dy = y[i+1] - y[i]
        dx = durations[i+1] - durations[i]
        slope = dy / dx
        slopes.append(slope)
    return pd.Series(slopes, index=slope_feature_names)

# Apply row-wise and join the new slope columns
df_slope_features = df_filtered.apply(compute_local_slopes, axis=1)
df_filtered = df_filtered.join(df_slope_features)
```
